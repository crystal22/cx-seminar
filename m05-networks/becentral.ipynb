# Consistency test

```python
for name in firstnames:
    edges = []
    for line in open(name + '.txt'):
        # should start with the first name
        if not line.startswith(name):
            print(name, 'is naughty!')
        
        edges.append(tuple(line.strip().split()))

    # multiple edge test
    if len(edges) != len(set(edges)):
        print('{}, there are multiple edges..'.format(name))
        from collections import defaultdict
        cnt = defaultdict(int)
        for x in edges:
            cnt[x] += 1
        for x in cnt:
            if cnt[x] > 1:
                print(x)
    
    # namespace test
    for u,v in edges:
        if u[:-1] not in firstnames or v[:-1] not in firstnames:
            print('{}, some names are wrong: {}'.format(name, u + ', ' + v ))
    
    # should be at most 20 lines
    if len(open(name + '.txt').readlines()) > 20:
        print(name, 'is naughty!')
```

# Reading edges

```python
import networkx as nx
G = nx.DiGraph()

for name in firstnames:
    G.add_edges_from([x.strip().split() for x in open(name + '.txt').readlines()])
```

```python
print(nx.info(G))
```

# In-degree

```python
names = set()
for n, deg in sorted(G.in_degree().items(), key=lambda x: x[1], reverse=True):
    name = n[:-1]
    if name in names:
        continue
    names.add(name)
    print('{:<10} {:<10}'.format(name, deg))
```

# PageRank

```python
names = set()
pr = nx.pagerank(G)
for n, p in sorted(pr.items(), key=lambda x: x[1], reverse=True):
    name = n[:-1]
    if name in names:
        continue
    names.add(name)
    print('{:<10} {:f}'.format(name, p))
```

# Betweenness centrality

```python
names = set()
betw = nx.algorithms.betweenness_centrality(G)
for n, b in sorted(betw.items(), key=lambda x: x[1], reverse=True):
    name = n[:-1]
    if name in names:
        continue
    names.add(name)
    print('{:<10} {:f}'.format(name, b))
```

# Closeness centrality

```python
names = set()
cl = nx.algorithms.closeness_centrality(G)
for n, c in sorted(cl.items(), key=lambda x: x[1], reverse=True):
    name = n[:-1]
    if name in names:
        continue
    names.add(name)
    print('{:<10} {:f}'.format(name, c))
```

```python
def draw_subg(name):
    g = nx.DiGraph()
    for firstname in firstnames:
        for x in open(firstname + '.txt'):
            n1, n2 = x.strip().split()
            if name in n1 or name in n2:
                g.add_edge(n1, n2)
    plt.figure(figsize=(8,8))
    nx.draw_spring(g, with_labels=True)
    
def draw_subg_only(name):
    g = nx.DiGraph()
    for x in open(name + '.txt'):
        n1, n2 = x.strip().split()
        g.add_edge(n1, n2)
    plt.figure(figsize=(8,8))
    nx.draw_spring(g, with_labels=True)

        
```
